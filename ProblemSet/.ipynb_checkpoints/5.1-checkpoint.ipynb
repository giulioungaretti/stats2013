## Plottin the data

```python
bins = range(-6,6,1)
```

```python
N = array([1,11,43,64,86,115,99,99,76,41,18,1])
Nerror = sqrt(N)
```

```python
fig,ax = subplots(figsize=(8,6))
ax.bar(bins,N,width=1,color ='#BBCCEE',alpha=0.9,label='Data',yerr=Nerror)
ax.set_ylabel('Count')
ax.set_xlabel('Some proprieties of the photons')
rstyle(ax)
```

Binned mean:

```python
[bins[i]+.5 for i in range(0,12)]
```

```python
mu = sum([N[i]*(bins[i]+.5) for i in range(0,12)]) * 1 / sum(N)
print  mu 
```

Binned error on the mean...

```python
std= sqrt((1./sum(N) *sum([N[i]*bins[i]**2 for i in range(0,12)])  - (1./sum(N)*sum([N[i]*bins[i] for i in range(0,12)]))**2))
print std
error_mu = std/sqrt(sum(N))
print error_mu
```

Calculate % error:

```python
(error_mu/mu) * 100
```

# Fitting with ROOT

Make histograms.

```python
canvas = rootnotes.default_canvas('canvas')
Hist = rt.TH1F('Hist', 'Hist', 12, -6.0, 6.0)
```

```python
for b,n in zip(range(1,13), N):
    Hist.SetBinContent(b,n)
Hist.Draw('E3')
```

Fit Gaussian

```python
fit_Sum = rt.TF1("fit_Sum", "gaus", -6, 6)
fit_Sum.SetLineColor(2)               # Set the line color to red.
fit_Sum.SetLineWidth(2)  
Hist.Fit("fit_Sum") #gaussian 
Hist.Draw('B')
Hist.Draw('E1 SAME')
canvas
```

Probability sucks! I think the distribution is skwed so I will fit with a skew normal distribution. Assuming a retarded detector that preferers brunettes.

```python
bgf =  rt.TF1("bgf","[0]*exp(-0.5*pow((x-[1])/([2]+(x>[1])*[3]*(x-[1])),2))",-6,6)
bgf2 =  rt.TF1("bgf2","[0]*exp(-0.5*pow((x-[1])/([2]+(x<[1])*[3]*(x-[1])),2))",-6,6)
```

Set reasonable paramter

```python
bgf.SetParameter(0,100) #scale factor
bgf.SetParameter(1,Hist.GetMean()/4) # mean
bgf.SetParameter(2,Hist.GetRMS()/4) # sdt
bgf.SetParameter(3,.5)   #skewness
bgf.SetLineColor(8)
bgf2.SetParameter(0,100) #scale factor
bgf2.SetParameter(1,Hist.GetMean()/4) # mean
bgf2.SetParameter(2,Hist.GetRMS()/4 ) # sdt
bgf2.SetParameter(3,.5)   #skewness
bgf2.SetLineColor(5)
```

Add fits

```python
Hist.Fit("bgf",'+');
Hist.Fit("bgf2","+");
```

```python
canvas
```

Get the probabilities of the fit:

try wiht poisson


```python
f1 = rt.TF1("f1","[0]*TMath::Poisson(x+[1],[2])",-6,6);
f1.SetParameter(0,5);
f1.SetParameter(1,-1);
f1.SetParameter(2,1);
f1.SetLineColor(6)
f1.Draw()
Hist.Fit('f1','+')
```

```python
canvas
```

use polynomial

```python
pol= rt.TF1("pol",'[0] + [1]*x**2 + [2]*x**4',-6,6);
pol.SetLineColor(41)
pol.Draw()
Hist.Fit('pol','+')
Hist.Draw()
canvas
```

get the probabilities of the fit:

```python
p_ske1 = bgf.GetProb()
p_pol = pol.GetProb()
p_gauss = fit_Sum.GetProb()
p_poiss = f1.GetProb()
print p_ske1
print p_pol
print p_gauss
print p_poiss
```

## KS test with ROOT

```python
fig2,ax2 = subplots(figsize=(8,6))

y1_sorted = np.sort(y1, kind='mergesort')
y2_sorted = np.sort(y2, kind='mergesort')
y3_sorted = np.sort(y3, kind='mergesort')
y4_sorted = np.sort(y4, kind='mergesort')
N_sorted = np.sort(N, kind='mergesort')


ax2.plot(x,y1_sorted,label='Skew')#p=%.e'%p_ske1)
ax2.plot(x,y2_sorted,label='Polynomial')# p=%.e'%p_ske2)
ax2.plot(x,y3_sorted,label='Gaussian')# p=%.e'%p_gauss)
ax2.plot(x,y4_sorted,label='Poisson')# p=%.e'%p_poiss)
ax2.plot(bins,N_sorted,label='Data')# p=%.e'%p_poiss)



pKS1 = rt.TMath.KolmogorovTest(len(bins), N_sorted, len(y1_sorted), y1_sorted,"")
print 'probability skew 1 +%.3e'%pKS1

aaaa = rt.TMath.KolmogorovTest( len(y2_sorted), y2_sorted ,len(bins), N_sorted ,"")
print 'probability Polynomial 2 +%.3e'%aaaa

pKS3 = rt.TMath.KolmogorovTest(len(bins), N_sorted, len(y3_sorted), y3_sorted,"")
print 'probability Gauss 1 +%.3e'%pKS3

pKS4 = rt.TMath.KolmogorovTest(len(bins), N_sorted, len(y4_sorted), y4_sorted ,"")
print 'probability Poisson 1 +%.3e'%pKS4


##sanity check test with itself

pKS3 = rt.TMath.KolmogorovTest(len(y2_sorted), y2_sorted, len(x), y3_sorted, "")
print  'probability gaussian and polynomial  1 +%.e'%pKS3


pKS3 = rt.TMath.KolmogorovTest(len(bins), N,len(bins), N,"")
print 'probability data with data 1 +%.e'%pKS3


pKS3 = rt.TMath.KolmogorovTest(len(y3), y3, len(y3), y3, "")
print  'probability gaussian with gaussian 1 +%.e'%pKS3
ax2.set_ylabel('Count')
ax2.set_xlabel('Sorted data.')

rstyle(ax2)
ax2.set_xlim(xmax=5)
ax2.legend(loc=0)
```
