## Monte Carlo simulation

```python
runs = [10,100,10000,100000]
fig,ax = subplots(figsize=(8,6))
prob = {}

for n_experiments in runs:
    vall = 0
    faces = 6
    tries =  10
    temp = [sum([random.randint(1,faces) for i in range(tries)]) for k in range(n_experiments)]
    
    faces = 10
    tries =  3
    temp2 = [sum([random.randint(1,faces) for i in range(tries)]) for k in range(n_experiments)]
    
    faces = 8
    tries =  6
    temp3 = [sum([random.randint(1,faces) for i in range(tries)]) for k in range(n_experiments)]
    
    
    vall =[i+j+k for i,j,k in zip (temp,temp2,temp3)]
    
    count = 0
    for i in vall:
        if i >= 100:
            count += 1  
    prob[n_experiments] = count
    print (float(count)/ n_experiments )*100
    hist(vall,normed=True, bins=20, alpha=.35, label ='Runs: %.e'%(n_experiments));

ax.legend(loc=0)
rstyle(ax)
```

```python
ax.set_xlabel('Sum of dice\' eyes')
ax.set_ylabel('Normalized frequency')
```
