```python
import ROOT as rt
import rootnotes 
from style import rstyle
```

```python
x = linspace(.1,1.4,14)
```

```python
y =  array([.19,.32,.43,.60,.73,.85,.98,1.08,1.11,1.15,1.22,1.22,1.27,1.34] )
y_erorr = 0.02 * np.ones(len(y))
```

```python
fig,ax = subplots()
ax.errorbar(x,y, yerr=y_erorr, fmt='.',label='data',color='#2DBCEA')
ax.set_xlabel('Some data')
ax.set_ylabel('Some values')
ax.legend(loc=0)
rstyle(ax)
```

## Test  linear fit

```python
canvas = rootnotes.default_canvas('asd')
```

```python
rt.gStyle.SetOptStat("emr")
rt.gStyle.SetOptFit(1111)
```

```python
Graph_x = rt.TGraphErrors(len(y), x, y, np.zeros(len(y)), y_erorr)
Graph_x.Draw('AP')
```

```python
fit_lin = rt.TF1("fit_lin", '[0]*x +[1]' , 0, 2)   # Here you need to define number of variables
fit_lin.SetParameters(-1.0, -1.0, 1.0)           # Remember to give good starting values!
fit_lin.SetLineColor(22)
Graph_x.Fit("fit_lin",'+')
Graph_x.Draw('APE')
```

```python
canvas
```

## ROOT Fit with kink

```python
# Defining external function:
def func_advanced (x, p) :
    x_t = .75
    if (x[0] < x_t) :
        return p[0]+p[1]*x[0]
    else :
        return p[2]+p[3]*x[0]



fit_x3 = rt.TF1("fit_x3", func_advanced, 0.3, 2.4, 4)   # Here you need to define number of variables
fit_x3.SetParameters(-1.0, -1.0, 1.0,1.0)           # Remember to give good starting values!
fit_x3.SetLineColor(3)
Graph_x.Fit("fit_x3", "+")
```

```python
canvas
```

```python
fit_x3.GetProb()
```

## Plotting

```python
y1 = np.array([fit_x3.Eval(float(i)) for i in x ])
ax.plot(x,y1,label='kined fit p = %.3f'%fit_x3.GetProb(),color='#F65C00')  # negative
y2 = np.array([fit_lin.Eval(float(i)) for i in x ])  # negative
ax.plot(x,y2,label='linear fit p = %.e'%fit_lin.GetProb(), color='#A2D400')  # negative
```
