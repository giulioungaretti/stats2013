```python
fig,ax = subplots()
t0_list  = [1,2,3]
tau_list = [2,7,14]
colors = ['#348ABD', '#A60628', '#7A68A6', '#467821','#D55E00',  '#CC79A7', '#56B4E9', '#009E73', '#F0E442', '0072B2']
colors = colors[0:-1:2]
for tau, t0, c in zip(tau_list, t0_list,colors):
    t= np.linspace(t0_list[0],30)
    C = exp( t0/tau )/tau
    f = 0
    f = C * exp(-t/tau)
    ax.plot(t,f,'-',color=c,label='C=%.2f'%(C))
    mu = float(t0+tau)
    ax.axvline(x= mu,ls='--',color=c)
    sigma = sqrt(abs(tau **2 - 4 * tau *t0))
    ax.axvspan(mu-sigma, mu+sigma, facecolor=c, alpha=0.1)
ax.set_xlabel('t')
ax.set_ylabel('f(t)')
ax.legend(loc=0)

rstyle(ax)
```
