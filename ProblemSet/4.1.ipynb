# ATLAS TRT

The ATLAS Transition Radiation Tracker (TRT) is a particle detector, which can se- parate electrons from pions through the number of High Threshold (HT) signal. On a track which crosses 35 detector units (straws), the probability for each straw to yield a HT signal is 4.2% for pions and 22.6% for electrons.

 -  What distribution does the number of HT signals NHT follow for electrons?

Binomial! with p0= 22.6% and number of tries n = 35

```python
p_e = 22.6 /100 
p_p = 4.2 /100
n = 35
```

```python
def binominalp(n,p,r):
	binomial_factor = factorial(n) / (factorial(r) *  factorial(n-r) )
	prob = binomial_factor * p ** r * (1-p)**(n-r)
	return prob
```

```python
r = np.linspace(0, 34,35)
```

```python
fig,ax = subplots(figsize=(8,6))
ax.bar(r+.5,binominalp(n,p_e,r),width=1,color ='#88e874',label='Electrons through TRT \n (p = %s, n = %s)'%(p_e ,n))
ax.bar(r+.5,binominalp(n,p_p,r),width=1,color ='#ffcd8c',label='Pions through TRT \n (p = %s, n = %s)'%(p_p ,n ),alpha=0.6)
ax.set_xlabel('Number of HT ')
ax.set_ylabel(r'$P_{N_{HT}}(r)$')
ax.legend()
rstyle(ax)
fig.savefig('Latex/fig/trt.pdf')
ax.set_xlim(-1
            ,20)
fig.savefig('Latex/fig/trt.pdf')
```
