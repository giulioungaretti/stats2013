```python
def load_multiple_data(files,data_dir):
    data = pd.read_csv(data_dir+files[0],sep = None, header = None, skiprows=3,skipinitialspace=True); # dear god, terrible text files (probablly crap data!!!)
    data = data.dropna(axis = 1, how = 'all')
    data.columns = ['L30','err30','L2','err2']
    #print data.head()
    for i in  files[1:]:
        print i
        temp = pd.read_csv(data_dir+i,sep = None, header = None, skiprows=3,skipinitialspace=True) # dear god, terrible text files (probablly crap data!!!)
        temp = temp.dropna(axis=1,how = 'all')
        temp.columns = ['L30','err30','L2','err2']
        data = data.append(temp)
        #print data.head()
    return data    
```

weighted mean:

```python
sum_up = 0
sum_down = 0
for value, sigma in zip(data_all['L30'].dropna(),data_all['err30'].dropna()):
    sum_up += value/ (sigma ** 2)
    sum_down += 1 /  (sigma ** 2)

w_mean = sum_up/sum_down
err_mean = (1/sqrt(sum_down))
    
    
#print str(w_mean)+'±'+str(err_mean)
print 'weighed mean 30 cm ruler  = %.4f±%.4f'%(w_mean,err_mean)
```

```python
rt.gStyle.SetOptStat("eMR");   # See http://root.cern.ch/root/html/TStyle.html#TStyle:SetOptStat
rt.gStyle.SetOptFit(1111);     # See http://root.cern.ch/root/html/TStyle.html#TStyle:SetOptFit
```

```python
Hist_L30cm = rt.TH1F("Hist_L30cm", "Lengths estimates by 30cm ruler",      100, 0.0, 5.0);
Hist_30cm_RvsE  = rt.TH2F("Hist_30cm_LvsE",   "Length residual vs. uncertainty estimates 30 cm ", 100, 0.0, 1, 250, 0.0, 1);

Hist_L2m = rt.TH1F("Hist_L2m", "Lengths estimates by 2 m ruler",      100, 0.0, 5.0);
Hist_2m_RvsE  = rt.TH2F("Hist_2m_RvsE",   "Length residual vs. uncertainty estimates 2m ", 100, 0.0, 1, 250, 0.0, 1);
```

```python
data_all.head()
```

## 30 cm

```python
for i in data_all['L30']:
    Hist_L30cm.Fill(i)
    
for i in  range(len(data_all['L30'].values)):
        Hist_30cm_RvsE.Fill(
                            abs(data_all['L30'].values[i]-data_all['L30'].mean()),data_all['err30'].values[i]
                            )
```

```python
canvas1 = rootnotes.canvas('L30')
Hist_L30cm.GetXaxis().SetRangeUser(0.0, 5.0)       # From the histogram, we get the x-axis, and set the range.
Hist_L30cm.SetMinimum(0)                           # We also set the minimum.
Hist_L30cm.GetXaxis().SetTitle("Table length (m)") # We give the x-axis a label.
Hist_L30cm.GetYaxis().SetTitle("Frequency")        # We give the y-axis a label.
Hist_L30cm.SetLineColor(rt.kBlue)                     # Set the histogram line color to blue.
Hist_L30cm.SetLineWidth(2)                         # Set the histogram line width to two.
Hist_L30cm.Draw()
```

Define a resonable fitting function:

```python
fit_L30cm = rt.TF1("fit_L30cm", "gaus", 0.0, 5.0)     # Define a (predefined) Gaussian fitting function.
fit_L30cm.SetLineColor(rt.kRed)                       # Set the line color to red.
fit_L30cm.SetLineWidth(2)                          # Set tne line width to 2.
Hist_L30cm.Fit("fit_L30cm", "LR")                  # Note the two options: L for Likelihood and R for Range.
Hist_L30cm.Draw()# Draw the histogram
canvas1
```

```python
canvas2 = rootnotes.canvas('erro30')
Hist_30cm_RvsE.SetMarkerStyle(7)
Hist_30cm_RvsE.GetXaxis().SetTitle('Rresidual') # We give the x-axis a label.
Hist_30cm_RvsE.GetYaxis().SetTitle('Estimates')        # We give the y-axis a label.
Hist_30cm_RvsE.Draw()

canvas2
```
