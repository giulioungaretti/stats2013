```python
%pylab inline
from style import rstyle
import pandas as pd
```

```python
def norm(df):
    df_norm = (df - df.mean()) / (df.max() - df.min())
    return df_norm
```

Data loading

```python
data_folder = './data/'
```

We have to define accurately what is gpd and gini.

GDP at purchaser's prices is the sum of gross value added by all resident producers in the economy plus any product taxes and minus any subsidies not included in the value of the products. It is calculated without making deductions for depreciation of fabricated assets or for depletion and degradation of natural resources. Data are in current U.S. dollars. Dollar figures for GDP are converted from domestic currencies using single year official exchange rates. For a few countries where the official exchange rate does not reflect the rate effectively applied to actual foreign exchange transactions, an alternative conversion factor is used.

```python
#world bank https://datamarket.com/data/set/15c9/gdp-per-capita-current-us#q=
gdp =  pd.read_csv(str(data_folder)+'gdp-per-capita-current-us.csv',skip_footer=2, index_col=0)
```

```python
gini = pd.read_csv(str(data_folder)+'gini-index.csv',skip_footer=2, index_col=0)
```

Normalized data

```python
gdp = norm(gdp)
```

Slice datas

```python
t_low = 1960
t_high = 2010
DT = t_high - t_low
gini_filtered = gini[(gini.index >= t_low) & (gini.index <= t_high )]
gdp_filtered = gdp[(gdp.index >= t_low) & (gdp.index <= t_high )]
```

Make filtered dataset 

```python
sample = {}
for name_1 in gdp_filtered.columns:
    if name_1 in gini_filtered.columns:
        sample[name_1]= [sum(gdp_filtered[name_1].dropna())/ (DT),average(gini_filtered[name_1].dropna())]
```

## Loop over year

```python
colors = ['#348ABD','#A60628', '#7A68A6', '#467821', '#D55E00',  '#CC79A7', '#56B4E9', '#009E73', '#F0E442', '#0072B2' ]
```

```python
temp = {}

t_low = 1950
t_high = 1980 

fig,ax = subplots()
years_low = range(1980,2010,5)
years_h = range(1990,2020,5)

for t_low, t_high, c in zip(years_low, years_h,colors):
#     ax.text(.7,70, str(t_low)+str(t_high) ,color = c)
    DT = t_high - t_low
    gini_filtered = gini[(gini.index >= t_low) & (gini.index <= t_high )]
    gdp_filtered = gdp[(gdp.index >= t_low) & (gdp.index <= t_high )]     
        
    sample = {}
    
    for name_1 in gdp_filtered.columns:
        if name_1 in gini_filtered.columns:
            sample[name_1]= [sum(gdp_filtered[name_1].values)/ (DT),average(gini_filtered[name_1].dropna())]
            
    for name in sample.keys():
        line = ax.plot(sample[name][0],sample[name][1],'o',color = c)
    line[0].set_label('range:'+str(t_low)+'- ' +str(t_high))
ax.legend( loc = 'upper right', bbox_to_anchor = (1.5, 1))
ax.set_xlabel('Integrated normalized GDP')
ax.set_ylabel('GINI index')
rstyle(ax)
```
