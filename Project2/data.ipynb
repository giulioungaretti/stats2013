# GDP and GINI index anaylsis

```python
%pylab inline
from style import rstyle
import pandas as pd
```

```python
def norm(df):
    df_norm = (df - df.mean()) / (df.max() - df.min())
    return df_norm
```

Data loading

```python
data_folder = './data/'

gdp = pd.read_csv(str(data_folder)+'gdp-current-us.csv',skip_footer=2, index_col=0)

gini = pd.read_csv(str(data_folder)+'gini-index.csv',skip_footer=2, index_col=0)
```

Normalized data

```python
gdp.Afghanistan.plot()
```

```python
gdp = norm(gdp)
```

Slice datas

```python
t_low = 1950
t_high = 1980
DT = t_high - t_low

gini_filtered = gini[(gini.index >= t_low) & (gini.index <= t_high )]
gdp_filtered = gdp[(gdp.index >= t_low) & (gdp.index <= t_high )]
```

Make filtered dataset 

```python
sample = {}

for name_1 in gdp_filtered.columns:
    if name_1 in gini_filtered.columns:
        sample[name_1]= [sum(gdp_filtered[name_1].values)/ (DT),average(gini_filtered[name_1].dropna())]
        
```

Plot it

```python
fig,ax = subplots()

for name in sample.keys():
    ax.plot(sample[name][0],sample[name][1],'o',label=name)
rstyle(ax)
```

## Loop over year

```python
colors = ['#348ABD','#A60628', '#7A68A6', '#467821', '#D55E00',  '#CC79A7', '#56B4E9', '#009E73', '#F0E442', '#0072B2' ]
```

```python
t_low = 1950
t_high = 1980 

fig,ax = subplots()
years_low = range(1950,2000,5)
years_h = range(1960,2020,5)

for t_low, t_high, c in zip(years_low, years_h,colors):
    DT = t_high - t_low
    gini_filtered = gini[(gini.index >= t_low) & (gini.index <= t_high )]
    gdp_filtered = gdp[(gdp.index >= t_low) & (gdp.index <= t_high )]     
        
    sample = {}
    
    for name_1 in gdp_filtered.columns:
        if name_1 in gini_filtered.columns:
            sample[name_1]= [sum(gdp_filtered[name_1].values)/ (DT),average(gini_filtered[name_1].dropna())]
            
    for name in sample.keys():
        ax.plot(sample[name][0],sample[name][1],'o',color = c)
ax.legend()
rstyle(ax)
```
